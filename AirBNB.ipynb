```python
import pandas as pd
data = pd.read_csv('/Users/krishisrani/Downloads/AB_NYC_2019.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
min_thresold, max_thresold = data.price.quantile([0.001,0.999])
min_thresold, max_thresold
                                                                     
```

```python
data[data.price < min_thresold]
```

```python
data[data.price > max_thresold]
```

```python
data = data[(data.price < max_thresold)&(data.price > min_thresold)]
data.shape
```

```python
# Every entry that contained rices that were not within the 0.1 and 99.9 percentiles of the data set were removed to clean any anomolies.
```

```python
new_df = data.groupby(['neighbourhood']).filter(lambda x: len(x) > 5)
#This filters out any neighborhoods with 5 or less entries
```

```python
new_df.groupby(['neighbourhood'])['price'].mean().nlargest()
#This gives the Top 5 neighborhoods based on the mean price in that neighborhood
```

```python
new_df.groupby(['neighbourhood'])['price'].mean().nsmallest()
#This gives the Bottom 5 neighborhoods based on the mean price in that neighborhood
```

```python
graph = data.groupby('neighbourhood_group')
```

```python
graph = graph[['neighbourhood_group','price']]
```

```python
totals = graph.mean()
```

```python
totals.sort_values(by = 'price')
```

```python
my_plot = totals.plot(kind='bar')
#This visualizes the price variation between neighborhood groups, showing that on average, Manhattan's prices are the greatest.
```
